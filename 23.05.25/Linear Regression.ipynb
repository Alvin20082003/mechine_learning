```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('J:\ml learning\23.05.25\Salary_Data[1].csv ')
```

```python
x = data[['Experience Years']].values  
y = data['Salary'].values             
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
numerator = np.sum((x - x_mean) * (y - y_mean))
denominator = np.sum((x - x_mean) ** 2)
m = numerator / denominator
b = y_mean - m * x_mean
y_pred = m * x + b
```

```python
print( m)
print( b)
print( y_pred)
```

```python
plt.scatter(x, y, color='blue', label='Actual Data')   # scatter plot of real data
plt.plot(x, y_pred, color='red', label='Regression Line')  # regression line
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression - Salary vs Experience')
plt.legend()
plt.grid(True)
plt.show()
```
